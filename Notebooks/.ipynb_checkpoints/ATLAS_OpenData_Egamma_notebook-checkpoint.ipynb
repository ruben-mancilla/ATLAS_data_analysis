1. We can start creating a [TChain](https://root.cern.ch/doc/master/classTChain.html) for each type of samples to be used. in this way we can add more the one file if they correspond the same family. For example:
    - A [TChain](https://root.cern.ch/doc/master/classTChain.html) for the real data with electrons/photos filter, so-called: **Egamma**
    - ggggg



```c++
TChain *InputDataEgamma = new TChain("mini");
TChain *InputDataMuons = new TChain("mini");
TChain *InputMonteCarlo =new TChain("mini");
```

Now this [datasets](http://opendata.atlas.cern/release/samples/Data/DataEgamma.root) correspond the same family as in the sted before:

```c++
InputDataEgamma->Add("/home/rufisica/Documentos/Cevale2_VE/Repositories/Dataset/Data/DataEgamma.root")
```

Ahora declaramos las variables que van hacer utilizadas para acceder a las variables de interes en el datasets.

```c++
Int_t InputEgamma_lep_n; 

Float_t InputEgamma_lep_pt[5],
        InputEgamma_lep_eta[5],
        InputEgamma_lep_phi[5],
        InputEgamma_lep_type[5]
```

Ya creadas las variables, ahora vamos a dar la dirección en la cual van a extraer la información cada una de ellas.

```c++
InputDataEgamma->SetBranchAddress("lep_n", &InputEgamma_lep_n);
InputDataEgamma->SetBranchAddress("lep_pt", &InputEgamma_lep_pt);
InputDataEgamma->SetBranchAddress("lep_phi", &InputEgamma_lep_phi);
InputDataEgamma->SetBranchAddress("lep_type", &InputEgamma_lep_type);
InputDataEgamma->SetBranchAddress("lep_eta", &InputEgamma_lep_eta);
```

Ahora vamos a crear los histogramas sobre el cual vamos a dar los resultados del análisis. Para mayor información de como trabajar con los mismo ir al siguiente enlace [https://docs.google.com/presentation/d/189f0qsDEnMSk2R5KWLRPz2TdEV5kTfXH1VcuAra4cnU/edit#slide=id.g29f3459464_2_531](https://docs.google.com/presentation/d/189f0qsDEnMSk2R5KWLRPz2TdEV5kTfXH1VcuAra4cnU/edit#slide=id.g29f3459464_2_531)

```c++
TH1F *h_lep_pt_1 = new TH1F("h_lep_pt_1", "Moment transverse of lepton 1;p_{t} [GeV];# Events",100,0,150);
TH1F *h_lep_pt_2 = new TH1F("h_lep_pt_2","Moment transverse of lepton 2;p_{t} [GeV];# Events",100,0,150);
TH1F *h_lep12_IM= new TH1F("h_lep12_IM","Invariant mass of the two first leptons;M_{ll} [GeV];# Events",100,0,150);
TH1I *h_lep12_type = new TH1I("h_lep12_type","h_lep12_type;Type;# Events",40,-20,20);
```

```c++
h_lep_pt_1->SetFillColor(kRed);
h_lep_pt_1->SetMarkerSize(0.5);
h_lep_pt_1->SetLineColor(0);

h_lep_pt_2->SetFillColor(kRed);
h_lep_pt_2->SetMarkerSize(0.5);
h_lep_pt_2->SetLineColor(0);

h_lep12_IM->SetFillColor(kBlue);
h_lep12_IM->SetMarkerSize(0.5);
h_lep12_IM->SetLineColor(1);
```

```c++
int nentries, nbytes, i ;
nentries=(Int_t)InputDataEgamma->GetEntries();

float M; // Invarriant mass of the two leading leptons

for (i=0;i<nentries;i++){
    nbytes = InputDataEgamma->GetEntry(i);
      
    // cut histogram
    if (InputEgamma_lep_n>1 && InputEgamma_lep_pt[1]>25000){
        
        h_lep12_type->Fill(InputEgamma_lep_type[0]); // Tipo de lepton seleccionado
        h_lep_pt_1->Fill( InputEgamma_lep_pt[0]/1000); // Momento trasnverso del lepton_{1}
        h_lep_pt_2->Fill(InputEgamma_lep_pt[1]/1000);  // Momento transverso del lepton_{2}
        
        // Calculation of the invariant mass of the two leading leptons, using:
        // InvMass^2=2.pt_{lep1}.pt_{2}.(\cosh(eta_{lep1}.eta_{lep2})-\cos(phi_{lep1}.phi_{lep2}))
        M=sqrt(2*InputEgamma_lep_pt[0]*InputEgamma_lep_pt[1]*
              ( TMath::CosH(InputEgamma_lep_eta[0]-InputEgamma_lep_eta[1])
              - TMath::Cos(InputEgamma_lep_phi[0]-InputEgamma_lep_phi[1])));
        
        // Fill
        h_lep12_IM->Fill(M/1000);
        
    }
       
}
```

Ahora vamos a crear las Canvas para poder visualizar los histogramas.

```c++
TCanvas *c1 = new TCanvas("c1","c1",50,50,800,600);
c1->SetFillColor(0);

TCanvas *c2 = new TCanvas("c2","c2",50,50,800,600); 
c2->SetFillColor(0);

TCanvas *c3 = new TCanvas("c3","c3",50,50,800,600);
c3->SetFillColor(0);

TCanvas *c4 = new TCanvas("c4","c4",50,50,800,600);
c4->SetFillColor(0);
```

Histograma del tipo de lepton.

```c++
c1->cd();
 h_lep12_type->Draw();
c1->Draw();
```

Histograma del $Lepton_1$

```c++
c2->cd();
h_lep_pt_1->Draw();
c2->Draw();
```

Histograma del Lepton_2.

```c++
c3->cd();
h_lep_pt_2->Draw();
c3->Draw();
```

Histograma de la masa invariante de los dos leptones.

```c++
c4->cd();
h_lep12_IM->DrawClone();
c4->Draw();
```

**Figura 3:** Masa invariante (*InvMass*) del sistema di-lepton. Para este calculo usamos la ecuación descrita en [Wikipedia:](https://en.wikipedia.org/wiki/Invariant_mass#Collider_experiments)
$$InvMass^2=2.pt_{lep1}.pt_{lep2}.(\cosh(eta_{lep1}-eta_{lep2})-\cos(phi_{lep1}-phi_{lep2}))$$

Ahora vamos Hacer un Fit del histograma.

```c++
c4->cd();
TF1 *g = new TF1("g","gaus",0,150);
g->SetLineColor(kRed);
g->SetLineStyle(2);
h_lep12_IM->Fit(g);
g->DrawClone("Same");
c4->Draw();
```
